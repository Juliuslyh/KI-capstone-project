```python
import pandas as pd
import numpy as np
import re
```

```python
#change the path for every csv
nameList=["time",'cont']
Data = pd.read_csv("result_2019_12.csv",names=nameList,sep=',',encoding='utf8')
```

```python
Data.shape
```

```python
Data.head()
```

```python
for i in range(0,len(Data['cont'])):
    Data['cont'][i] = re.sub(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*', '', Data['cont'][i])
    Data['cont'][i] = Data['cont'][i].replace("#"," ")
    Data['cont'][i] = Data['cont'][i].replace("\n","")
```

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer

german_stop_words = stopwords.words('german')
```

```python
def stop_word_removal(x):
    token = x.split()
    return ' '.join([w for w in token if not w in german_stop_words])
```

```python
#remove stopwords
tweet = Data['cont'].apply(stop_word_removal)
tweet = pd.DataFrame(tweet)
```
