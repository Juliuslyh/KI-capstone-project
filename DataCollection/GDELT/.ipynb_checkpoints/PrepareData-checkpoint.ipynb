```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from pathlib import Path
from datetime import datetime
```

```python
absatz_csv = Path('/content/drive/My Drive/AI Capstone/Daten(Absatz, Bewertung, Wetter und Feiertagen )/Bierabsatz.csv')
tweets_csv = Path('/content/drive/My Drive/AI Capstone/Ready for Analysis/Hessen_2015_2020_percentages_of_pntweets.csv')
weather_csv = Path('/content/drive/My Drive/AI Capstone/Daten(Absatz, Bewertung, Wetter und Feiertagen )/Weather_Frankfurt.csv')
holiday_csv = Path('/content/drive/My Drive/AI Capstone/Daten(Absatz, Bewertung, Wetter und Feiertagen )/date_holiday_monthly.csv')
gdelt_folder = Path('/content/drive/My Drive/AI Capstone/GDELT/Data/')
out_folder = Path('/content/drive/My Drive/AI Capstone/Ready for Analysis/')
```

```python
month_dict = {
    "Januar" : 1,
    "Februar" : 2,
    "März" : 3,
    "April" : 4,
    "Mai" : 5,
    "Juni" : 6,
    "Juli" : 7,
    "August" : 8,
    "September" : 9,
    "Oktober" : 10,
    "November" : 11,
    "Dezember" : 12
}
```

```python
# load data for one state and optimize the structure
# calculate the difference per month to make the time series stationary
def load_bundesland_absatz(file, bundesland, month_dict):
  absatz = pd.read_csv(file)
  df = absatz[absatz['region'].str.contains(bundesland, na=False, case=False)].copy()
  df[['year', 'month']] = df.time.str.split("-", expand=True)
  col = df.columns[0]
  df.drop(columns=[col],axis=1,inplace=True)
  df.reset_index(drop=True, inplace=True)
  df = df.iloc[::-1]
  df.reset_index(drop=True, inplace=True)
  df['month'] = df['month'].str.strip().map(month_dict)
  df['diff'] = df['absatz(hl)'].diff()
  df.dropna(inplace=True)
  display(df.head())
  plt.plot(df['time'], df['absatz(hl)'])
  plt.ylabel('Absatz (Hl)')
  plt.xlabel('Time')
  plt.show()
  plt.plot(df['time'], df['diff'], 'r')
  plt.ylabel('Absatz Diff')
  plt.xlabel('Time')
  plt.show()
  return df

absatz_hessen = load_bundesland_absatz(absatz_csv, 'Hessen', month_dict)
```

```python
def load_tweets(file):
  df = pd.read_csv(file)
  df[['year', 'month']] = df.Year_Month.str.split("_", expand=True)
  display(df.head())
  plt.plot(df['Year_Month'], df['Tweet_Counts'])
  plt.ylabel('Tweet_Counts')
  plt.xlabel('Time')
  plt.show()
  plt.plot(df['Year_Month'], df['Positive_Percentage(%)'], 'g',
           df['Year_Month'], df['Negative_Percentage(%)'], 'r')
  plt.ylabel('Percentages')
  plt.xlabel('Time')
  plt.show()
  return df

tweets_hessen = load_tweets(tweets_csv)
```

```python
def load_weather(file):
  df = pd.read_csv(file)
  df.rename( columns={'Unnamed: 0':'month_year',}, inplace=True )
  df[['month', 'year']] = df.month_year.str.split("-", expand=True)
  df['Temperature-High'] = df['Temperature-High'].str.split().str[0]
  df['Temperature-Low'] = df['Temperature-Low'].str.split().str[0]
  df['Temperature-Average'] = df['Temperature-Low'].str.split().str[0]
  df['Humidity-High'] = df['Humidity-High'].str.split('%').str[0]
  df['Humidity-Low'] = df['Humidity-Low'].str.split('%').str[0]
  df['Humidity-Average'] = df['Humidity-Low'].str.split('%').str[0]
  df['Pressure-High'] = df['Pressure-High'].str.split().str[0]
  df['Pressure-Low'] = df['Pressure-Low'].str.split().str[0]
  df['Pressure-Average'] = df['Pressure-Low'].str.split().str[0]
  
  display(df.head())
  plt.plot(df['month_year'], df['Temperature-Average'])
  plt.ylabel('Temperature')
  plt.xlabel('Time')
  plt.show()
  return df

weather_hessen = load_weather(weather_csv)
```

```python
def load_holiday(file):
  df = pd.read_csv(file)
  df.rename( columns={'Unnamed: 0':'year_month',}, inplace=True )
  df[['year', 'month']] = df.year_month.str.split("-", expand=True)
  df['month'] = df['month'].astype('int32')
  display(df.head())
  plt.plot(df['year_month'], df['num of Non-working day'])
  plt.ylabel('Non-working days')
  plt.xlabel('Time')
  plt.show()
  return df

holiday_hessen = load_holiday(holiday_csv)
```

```python
def prepare_gdelt_year(folder, year):
  full_stat = []
  for i in range(1, 13):
    try:
      data_path = os.path.join(folder, f'brauerei_{year}_bereinigt/brauerei_{year}_{i}.csv')
      df = pd.read_csv(data_path, converters={'V2Tone': eval})
      v2_tone = df['V2Tone'].to_numpy()
      v2_tone = np.array([*v2_tone])
      tone = v2_tone[:, 0]
      polarity = v2_tone[:, 3]
      ratio = (tone < 0).sum()/(tone >= 0).sum()
      stat = np.array([np.mean(tone), np.std(tone), ratio,
                       len(tone), np.mean(polarity), np.std(polarity)])
      full_stat.append(stat)
    except:
      pass
  full_stat = np.array(full_stat)
  df1 = pd.DataFrame(full_stat, columns = ['tone_mean', 'tone_std', 'ratio_pos_neg',
                                           'number_of_articles', 'polarity_mean',
                                           'polarity_std'])
  df1['year'] = year
  if year == 2015:
    df1['month'] = df1.index + 3
  else:
    df1['month'] = df1.index + 1
  return df1
```

```python
def load_gdelt():
  df = pd.DataFrame()
  for year in range(2015, 2021):
    df_year = prepare_gdelt_year(gdelt_folder, year)
    df = df.append(df_year)
  df['year_month'] = df['year'].astype(str) + '_' + df['month'].astype(str)
  display(df.head())
  plt.plot(df['year_month'], df['tone_mean'], 'b', df['year_month'], df['tone_std'], 'r')
  plt.ylabel('tone_mean')
  plt.xlabel('Time')
  plt.show()
  plt.plot(df['year_month'], df['ratio_pos_neg'])
  plt.ylabel('ratio')
  plt.xlabel('Time')
  plt.show()
  plt.plot(df['year_month'], df['number_of_articles'])
  plt.ylabel('articles')
  plt.xlabel('Time')
  plt.show()
  plt.plot(df['year_month'], df['polarity_mean'], 'b', df['year_month'], df['polarity_std'], 'r')
  plt.ylabel('polarity_mean')
  plt.xlabel('Time')
  plt.show()
  return df


gdelt = load_gdelt()
```
